```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('grayscale_image.jpg')

image = image / 255

gamma = 2.2
gamma_corrected = np.power(image, gamma)

gamma_corrected = np.uint8(gamma_corrected * 255)

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(gamma_corrected)
plt.title(f"Gamma Corrected (={gamma})")
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("grayscale_image.jpg")

negative_image = 255 - image 

plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(negative_image, cmap="gray")
plt.title("Negative Image")
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("grayscale_image.jpg")

image = image / 255.0  

c = 255 / np.log(1 + np.max(image)) 
log_transformed = c * np.log(1 + image)

log_transformed = np.uint8(log_transformed)

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(log_transformed)
plt.title("Log Transformed Image")
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = np.zeros((300, 300), dtype=np.uint8)
img2 = np.zeros((300, 300), dtype=np.uint8)

cv2.rectangle(img1, (50, 50), (200, 200), 255, -1)  
cv2.rectangle(img2, (100, 100), (250, 250), 255, -1)  

union = cv2.bitwise_or(img1, img2)          
intersection = cv2.bitwise_and(img1, img2) 
difference = cv2.bitwise_and(img1, cv2.bitwise_not(img2))  
xor = cv2.bitwise_xor(img1, img2)     

plt.figure(figsize=(3, 3))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img2, cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.tight_layout()
plt.show()

titles = ['Union', 'Intersection', 'Difference', 'XOR']
images = [union, intersection, difference, xor]

for i in range(4):
    plt.subplot(1, 4, i + 1)
    plt.imshow(images[i], cmap = "gray")
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
