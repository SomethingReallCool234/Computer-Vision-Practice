```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import color

image = cv2.imread("image_converted.jpg")  
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY) 

hog_features, hog_image = hog(gray, orientations=9, pixels_per_cell=(4, 4),
                              cells_per_block=(2, 2), visualize=True, block_norm='L2-Hys')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.imshow(image)
ax1.set_title("Original Image")
ax1.axis("off")

ax2.imshow(hog_image, cmap="gray")
ax2.set_title("HOG Features")
ax2.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('pedestrian.jpg')
img = cv2.resize(img, (800, 600))  # Optional resizing for speed

# Initialize HOG descriptor/person detector
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

# Detect people
(rects, weights) = hog.detectMultiScale(
    img,
    winStride=(8, 8),
    padding=(8, 8),
    scale=1.05
)

# Draw bounding boxes
for (x, y, w, h) in rects:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Convert to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Plot the image
plt.figure(figsize=(12, 8))
plt.imshow(img_rgb)
plt.title("Pedestrian Detection using HOG")
plt.axis('off')
plt.show()
```
