```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('grayscale_image.jpg')

scale_percent = 5
width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)
scaled_image = cv2.resize(image, (width, height))

angle = 45 
center = (image.shape[1] // 2, image.shape[0] // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1)
rotated_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]))

tx, ty = 150, 100
shift_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
shifted_image = cv2.warpAffine(image, shift_matrix, (image.shape[1], image.shape[0]))

titles = ['Original', 'Scaled', 'Rotated', 'Shifted']
images = [image, scaled_image, rotated_image, shifted_image]

for i in range(4):
    plt.subplot(1, 4, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('grayscale_image.jpg')

reflected_x = cv2.flip(image, 0)

reflected_y = cv2.flip(image, 1)

reflected_xy = cv2.flip(image, -1)

titles = ['Original', 'Reflection X', 'Reflection Y', 'Reflection XY']
images = [image, reflected_x, reflected_y, reflected_xy]

for i in range(4):
    plt.subplot(1, 4, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('grayscale_image.jpg')

shear_factor = 0.3
shear_matrix = np.float32([[1, shear_factor, 0], [0, 1, 0]])

height, width = image.shape[:2]
sheared_image = cv2.warpAffine(image, shear_matrix, (width + int(height * shear_factor), height))

plt.imshow(sheared_image)
plt.title('Sheared Image')
plt.axis('off')
plt.show()
```
